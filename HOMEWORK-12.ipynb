```python
import threading

# Function to check if a number is prime
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

# Function each thread will run
def find_primes_in_range(start, end, primes):
    for num in range(start, end):
        if is_prime(num):
            primes.append(num)

# Main program
if __name__ == "__main__":
    start_range = 1
    end_range = 100
    num_threads = 4

    # Divide the range into equal parts
    step = (end_range - start_range) // num_threads
    threads = []
    primes = []

    # Create and start threads
    for i in range(num_threads):
        start = start_range + i * step
        end = start + step if i < num_threads - 1 else end_range + 1
        thread = threading.Thread(target=find_primes_in_range, args=(start, end, primes))
        threads.append(thread)
        thread.start()

    # Wait for all threads to finish
    for thread in threads:
        thread.join()

    # Sort and display primes
    primes.sort()
    print("Prime numbers:", primes)
```

```python
import threading
from collections import Counter

# Function each thread will run
def count_words_in_lines(lines, word_count):
    local_count = Counter()
    for line in lines:
        words = line.lower().split()
        local_count.update(words)
    word_count.append(local_count)

# Main program
if __name__ == "__main__":
    filename = "large_text.txt"  # make sure the file exists
    num_threads = 4

    # Read file lines
    with open(filename, 'r', encoding='utf-8') as f:
        lines = f.readlines()

    # Divide lines among threads
    step = len(lines) // num_threads
    threads = []
    word_count_list = []

    for i in range(num_threads):
        start = i * step
        end = (i + 1) * step if i < num_threads - 1 else len(lines)
        thread = threading.Thread(target=count_words_in_lines, args=(lines[start:end], word_count_list))
        threads.append(thread)
        thread.start()

    for thread in threads:
        thread.join()

    # Combine all word counts
    total_count = Counter()
    for count in word_count_list:
        total_count.update(count)

    # Display the results
    print("Top 10 most common words:")
    for word, freq in total_count.most_common(10):
        print(f"{word}: {freq}")
```
