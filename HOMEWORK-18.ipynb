```python
df = pd.read_csv(r"C:\Users\Asus\Downloads\titanic.csv")
df.head()

female_20_30 = df[
    (df['Sex']=='female') &
    (df['Pclass'] == 1) &
    (df['Age'] >= 20) &
    (df['Age'] <= 30)
]
print(female_20_30)

fare_100  = df[df['Fare'] > 100]
print(fare_100)

survived_alone = df[
    (df['Survived'] == 1) &
    (df['SibSp'] == 0) &
    (df['Parch'] == 0)
]
print(survived_alone)

embarked = df[(df['Embarked'] == 'C') & (df['Fare'] > 50)]
print(embarked)

with_family = df[(df['SibSp']>0) & (df['Parch']>0)]
print(with_family)

children_not_survived = df[(df['Age']<=15) & (df['Survived']==0)]
print(children_not_survived)

high_fare_cabins = df[(df['Cabin'].notnull()) & (df['Fare']>200)]
print(high_fare_cabins)

odd_ids = df[(df['PassengerId'])%2==1]
print(odd_ids)

unique_tickets = df[(df['Ticket'].duplicated(keep=False) == False)]
print(unique_tickets)

miss_class1 = df[(df['Name'].str.contains('Miss')) & (df['Pclass']==1)]
print(miss_class1)
```
